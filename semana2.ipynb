# Análise exploratória de dados: distribuições e sumarização
Este notebook explora técnicas de visualização de distribuições e sumarização de dados usando o **Pandas** e **Matplotlib**.

A análise inclui:
- visualização de distribuições com **histogramas**;
- identificação de outliers e dispersão com **boxplots**;
- cálculo de estatísticas descritivas usando **tabelas sumarizadas**;
- transformação e análise de **variáveis categóricas**.

Este material funciona como referência para análise exploratória básica e entendimento do formato dos dados.

## Bibliotecas utilizadas
- `pandas`: manipulação e análise de dados estruturados.
- `numpy`: geração de dados aleatórios e cálculos numéricos.
- `matplotlib.pyplot`: controle visual dos gráficos.

## Conjunto de dados
Vamos criar um `pd.DataFrame` fictício contendo informações de **idade** e **salário** de 100 indivíduos, gerados de forma aleatória com `numpy`.

```python
np.random.seed(0)

dados = pd.DataFrame({
    'idade': np.random.randint(18, 70, 100),
    'salario': np.random.normal(5000, 1500, 100)
})

print(dados.head())
```

## Histogramas
O histograma ajuda a visualizar a **distribuição de frequências** de variáveis contínuas, agrupando os dados em intervalos (bins).

- Mostra a concentração de valores em determinadas faixas;
- Ajuda a identificar a forma da distribuição (ex: normal, assimétrica).

### Interpretação
O método `.hist()` do Pandas é uma forma rápida de plotar histogramas diretos de colunas do DataFrame, como `idade` e `salario`.

```python
dados.idade.hist(bins=10, color='skyblue', edgecolor='black')
```

```python
dados.salario.hist(bins=15, color='orange', edgecolor='black')
```

## Boxplots
O boxplot (gráfico de caixa) é excelente para visualizar a **dispersão**, **assimetria** e identificar **outliers** (valores atípicos).

- A caixa central representa os 50% centrais dos dados (entre o 1º e 3º quartil);
- A linha no meio da caixa é a **mediana**;
- Os "bigodes" estendem-se até os valores normais extremos, e pontos fora são outliers.

### Observações
A função `.boxplot()` do Pandas facilita a visualização rápida de estatísticas de dispersão para variáveis numéricas como o `salario`.

```python
dados.boxplot(column='salario')
```

## Tabelas sumarizadas
Estatísticas descritivas numéricas fornecem um resumo quantitativo rápido dos dados.

- Medidas como **média**, **mediana** e **desvio padrão** descrevem o centro e o espalhamento;
- A função `describe()` exibe um resumo completo com contagem, média, quartis e valores min/max de uma só vez.

### Interpretação
O `.describe()` é uma das ferramentas iniciais mais úteis no **Pandas** para entender as características numéricas de um dataset rapidamente.

```python
media = np.mean(dados.salario)
mediana = np.median(dados.salario)
desvio = np.std(dados.salario)

print(f'Média: {media}\nMediana: {mediana}\nDesvio Padrão: {desvio}')
```

```python
print(dados.describe())
```

## Variáveis categóricas
Muitas vezes é útil transformar variáveis contínuas em **categóricas** (agrupamento/binning) para análises comparativas.

- O agrupamento simplifica variáveis contínuas (como idade) em faixas qualitativas;
- Permite analisar as **frequências** de cada grupo criado.

### Observações
A função `pd.cut()` é usada para segmentar e ordenar valores em categorias. Depois, `.value_counts()` conta os registros em cada faixa.

```python
dados['categoria'] = pd.cut(dados.idade, bins=[17, 30, 50, 70], labels=['Jovem', 'Adulto', 'Idoso'])
frequencia = dados.categoria.value_counts()
print(frequencia)
```

## Conclusões
- Os **histogramas** e **boxplots** são ferramentas visuais essenciais para entender a distribuição e os limites dos dados numéricos.
- As **tabelas sumarizadas** complementam a visualização com valores numéricos exatos (média, quartis, etc.).
- Transformar dados contínuos em **variáveis categóricas** pode revelar novos padrões e simplificar análises baseadas em grupos.
