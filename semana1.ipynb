# Análise exploratória de dados
Este notebook apresenta conceitos fundamentais de **estatística descritiva** aplicados com **Python**.

A análise inclui:
- cálculo de **medidas de tendência central** (média, mediana e moda);
- cálculo de **medidas de posição** (quartis, decis e percentis);
- visualização de dados com **histograma**.

Este material funciona como referência para análise exploratória de dados e interpretação de estatísticas descritivas.

## Bibliotecas utilizadas
- `pandas`: manipulação de dados em tabelas e séries.
- `numpy`: operações numéricas e cálculo de percentis.
- `matplotlib.pyplot`: criação de gráficos e visualizações.

## Conjunto de dados
Vamos usar uma `pd.Series` com notas de alunos para calcular as medidas de tendência central, e um `np.array` com valores numéricos para explorar as medidas de posição.

## Medidas de tendência central
As medidas de tendência central resumem um conjunto de dados em um único valor representativo.

- **Média:** soma de todos os valores dividida pela quantidade. Calculada com `notas.mean()`.
- **Mediana:** valor central quando os dados estão ordenados. Calculada com `notas.median()`.
- **Moda:** valor que aparece com mais frequência. Calculada com `notas.mode()`.

### Observações
A média é sensível a **valores extremos** (outliers), enquanto a mediana é mais **robusta**. A moda pode não ser única.

```python
notas = pd.Series([7.5, 8.0, 6.0, 9.0, 8.0, 7.0])
print('Média: ', notas.mean())
print('Mediana: ', notas.median())
print('Moda: ', notas.mode()[0])
```

## Medidas de posição
As medidas de posição indicam onde um valor se situa dentro da distribuição dos dados.

- **Quartis:** dividem os dados em **4 partes iguais** (Q1, Q2, Q3). Calculados com `np.percentile(dados, [25, 50, 75])`.
- **Decis:** dividem os dados em **10 partes iguais**. Calculados com `np.percentile(dados, np.arange(10, 100, 10))`.
- **Percentis:** dividem os dados em **100 partes iguais**. O percentil 90 indica que 90% dos dados estão abaixo desse valor.

### Interpretação
A função `np.percentile()` do **NumPy** permite calcular qualquer percentil de um conjunto de dados, sendo a base para quartis e decis.

```python
dados = np.array([12, 15, 18, 19, 21, 22, 24, 25, 27, 30])

quartis = np.percentile(dados, [25, 50, 75])
print('Quartis: ', quartis)

decis = np.percentile(dados, np.arange(10, 100, 10))
print('Decis: ', decis)

percentil_90 = np.percentile(dados, 90)
print('Percentil 90: ', percentil_90)
```

## Histograma
O histograma é um gráfico de barras que representa a **distribuição de frequência** de uma variável numérica.

- O eixo **x** mostra os intervalos (bins) dos valores;
- O eixo **y** mostra a **frequência** de ocorrências em cada intervalo;
- Permite visualizar a **forma da distribuição** (simétrica, assimétrica, etc.).

### Observações
O método `.plot.hist()` do **Pandas** utiliza o `matplotlib` internamente. O parâmetro `bins` controla o número de intervalos do histograma.

```python
notas.plot.hist(bins=5, edgecolor='black')
plt.title('Notas dos alunos')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.show()
```

## Conclusões
- As **medidas de tendência central** (média, mediana e moda) resumem o comportamento geral dos dados.
- As **medidas de posição** (quartis, decis e percentis) indicam como os valores se distribuem.
- O **histograma** permite visualizar a distribuição de frequência de forma clara.
