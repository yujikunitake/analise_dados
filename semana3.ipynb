```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Análise multivariada: correlação e visualização multivariada
Este notebook explora as relações entre **peso**, **potência**, **consumo** e **preço** de carros.

A análise inclui:
- cálculo da matriz de correlação;
- visualização de mapa de calor;
- gráficos de dispersão (pairplot) para comparar pares de variáveis.

Este material funciona como referência para análise exploratória de dados e interpretação de correlações.

## Bibliotecas utilizadas
- `pandas`: manipulação de dados em tabelas.
- `seaborn`: criação de gráficos estatísticos.
- `matplotlib.pyplot`: controle de figuras e exibição de gráficos.

## Conjunto de dados
Vamos montar um DataFrame com exemplos de carros e suas características principais.

```python
dados = {
    'Peso': [1500, 1700, 1600, 1400, 1800],
    'Potência': [110, 130, 125, 100, 140],
    'Consumo': [14, 12, 13, 15, 11],
    'Preço': [20000, 25000, 24000, 22000, 27000]
}

df = pd.DataFrame(dados)
print(df)
```

## Matriz de correlação
A correlação de Pearson mede a relação linear entre duas variáveis.

- Valores próximos de **+1** indicam relação positiva forte;
- Valores próximos de **-1** indicam relação negativa forte;
- Valores próximos de **0** indicam pouca relação linear.

A matriz `df.corr()` mostra essas relações para todas as combinações de variáveis.

```python
correlacao = df.corr()
print(correlacao)
```

## Mapa de Correlação
Este mapa de calor mostra a força e a direção das relações entre as variáveis.

- **Vermelho:** correlação positiva
- **Azul:** correlação negativa

O peso tende a ter correlação negativa com consumo e preço, enquanto a potência tem correlação positiva com o preço.

### Observações
Potência e consumo mostram correlação moderada com preço. O gráfico ajuda a identificar quais variáveis são mais relevantes para análise e tomada de decisão.

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Calor da Correlação entre Variáveis do Carro')
plt.show()
```

## Pairplot: Relações entre variáveis
O pairplot visualiza a dispersão entre pares de variáveis e mostra a distribuição individual de cada variável na diagonal.

- **Diagonal:** histograma ou distribuição de cada variável
- **Fora da diagonal:** comparação entre pares de variáveis

### Interpretação
Use este gráfico para identificar padrões, tendências e possíveis correlações entre peso, potência, consumo e preço.

```python
sns.pairplot(df)
plt.suptitle('Gráficos de Dispersão para Pares de Variáveis', y=1.02)
plt.show()
```

## Conclusões
- O mapa de correlação facilita a identificação das relações lineares entre variáveis.
- O pairplot permite visualizar padrões e tendências entre pares de variáveis.
