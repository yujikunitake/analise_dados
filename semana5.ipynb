```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Tratamento de Dados: Valores Faltantes e Detecção de Outliers
Este notebook explora técnicas essenciais para a preparação de dados antes da modelagem, focando no **tratamento de valores faltantes** e na **identificação e tratamento de outliers** usando **Pandas**.

A análise inclui:
- identificação de dados ausentes (NaNs);
- técnicas de preenchimento (imputação) ou remoção de valores faltantes;
- detecção de outliers usando o método do Intervalo Interquartil (IQR);
- visualização de outliers com **boxplots**.

Este material funciona como referência para garantir a qualidade e a consistência da base de dados.

## Bibliotecas utilizadas
- `pandas`: manipulação e análise de dados estruturados.
- `numpy`: operações numéricas e geração de dados ausentes (`np.nan`).
- `seaborn` e `matplotlib.pyplot`: visualização de dados.

## Conjunto de dados
Vamos criar um `pd.DataFrame` fictício contendo informações de **idade** e **renda**, inserindo propositalmente alguns valores nulos e outliers para simular dados do mundo real.

```python
np.random.seed(42)

dados = pd.DataFrame({
    'idade': np.random.randint(18, 65, 50),
    'renda': np.random.normal(4000, 1000, 50)
})

# Inserindo valores faltantes (NaN)
dados.loc[5:10, 'idade'] = np.nan
dados.loc[20:25, 'renda'] = np.nan

# Inserindo outliers
dados.loc[48, 'renda'] = 25000
dados.loc[49, 'renda'] = 30000

print("Primeiras linhas do dataset:")
print(dados.head(15))
```

## Valores Faltantes
Identificar a quantidade de valores nulos e decidir a melhor estratégia de tratamento.

```python
print("Valores faltantes por coluna:")
print(dados.isnull().sum())
```

```python
# Tratamento 1: Preenchendo valores ausentes com a mediana
dados_tratados = dados.copy()
dados_tratados['idade'] = dados_tratados['idade'].fillna(dados_tratados['idade'].median())
dados_tratados['renda'] = dados_tratados['renda'].fillna(dados_tratados['renda'].median())

print("Valores faltantes após tratamento:")
print(dados_tratados.isnull().sum())
```

## Detecção e Tratamento de Outliers
Visualizando e detectando anomalias.

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=dados_tratados['renda'])
plt.title('Boxplot de Renda (Com Outliers)')
plt.show()
```

```python
# Método IQR (Intervalo Interquartil)
Q1 = dados_tratados['renda'].quantile(0.25)
Q3 = dados_tratados['renda'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrando os outliers
dados_sem_outliers = dados_tratados[(dados_tratados['renda'] >= limite_inferior) & (dados_tratados['renda'] <= limite_superior)]

print(f"Quantidade original: {len(dados_tratados)}")
print(f"Quantidade sem outliers: {len(dados_sem_outliers)}")
```
